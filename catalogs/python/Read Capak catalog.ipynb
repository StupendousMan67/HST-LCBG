```python
%matplotlib inline

from astropy.io import ascii    
import numpy as np
import matplotlib.pyplot as plt
```

```python
infile = '../Capak/deimos_redshifts.tbl'
```

```python
data = ascii.read( infile)
```

```python
data.info
```

select targets with high-quality spectra

```python
buf = np.where( (data['Q'] == 2.0))
good = buf[0]
valid = data[good]
n = len(valid)
print n
```

Show locations of high-quality spectra

```python
plt.figure(figsize=(14,8))
plt.plot( valid['ra'], valid['dec'], 'b,' )
```

select targets in range 0.5<z<0.6

```python
buf = np.where( (data['Q'] == 2.0) & (data['zspec'] > 0.5) & (data['zspec'] < 0.6))
good = buf[0]
valid = data[good]
n = len(valid)
print n
```

```python
plt.figure(figsize=(14,8))
plt.plot( valid['ra'], valid['dec'], 'r.' )
```
