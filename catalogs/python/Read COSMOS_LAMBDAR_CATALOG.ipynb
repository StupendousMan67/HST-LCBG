Determine the LCBG fraction.  Per Lucas, preferred selection criteria are: 
* SG_MASTER==0
  *  star-galaxy separation flag using information from COSMOS2015, our size-magnitude classification and visual inspection.  
  * Final classifications are as follows: 0=galaxy, 1=star, 2=x-ray point source, 3=x-ray extended source and -9999=no star-galaxy flag.  
  * This selects on GALAXIES based on morphology
* & Z_BEST>0.0001
  * Z_BEST       = Best redshift available selected by the method detailed by Davies+2015
  * This appears to select targets with available redshifts
* & Z_USE<3
  * Z_USE        = Redshift quality flag as defined by Davies+2015 (use Z_USE < 3 for reliable spectroscopic redshifts, Z_USE ==1 for reliable high-resolution spectroscopic redshift, Z_USE ==2 for reliable PRIMUS redshift)
  * So this is selecting for high-quality redshifts
* & LAMBDAR_COSMOS2015_sep<0.8
  * LAMBDAR_COSMOS20XX_sep = offset in position matching between catalogues. Note matches of >0.8'' may be unreliable.
  * This appears to select for targets that have the same position is different catalogs

```python
n_all = len(data)
```

```python
buf = np.where( data['SG_MASTER'] == 0)
good = buf[0]
lcbg = data[good]
n_lcbg = len(lcbg)
print n_lcbg
```

```python
buf = np.where( data['Z_BEST'] > 0.0001)
good = buf[0]
lcbg = data[good]
n_lcbg = len(lcbg)
print n_lcbg
```

```python
buf = np.where( data['Z_USE'] < 3)
good = buf[0]
lcbg = data[good]
n_lcbg = len(lcbg)
print n_lcbg
```

```python
buf = np.where( data['LAMBDAR_COSMOS2015_sep'] < 0.8)
good = buf[0]
lcbg = data[good]
n_lcbg = len(lcbg)
print n_lcbg
```

```python
buf = np.where( (data['SG_MASTER'] == 0) & (data['Z_BEST'] > 0.0001) & (data['Z_USE'] < 3) & (data['LAMBDAR_COSMOS2015_sep'] < 0.8))
good = buf[0]
lcbg = data[good]
n_lcbg = len(lcbg)
print n_lcbg
```

```python
lcbg_fraction = float(n_lcbg) / float(n_all)
print lcbg_fraction
```

Plot the positions of LCBG targets on sky

```python
plt.figure(figsize=(14,8))
plt.plot( lcbg['RA'], lcbg['DEC'], 'b,' )
```

Determine distribution of candidate comparison LCBGs in the 0.5 < z < 0.6 range.

```python
buf = np.where( (data['SG_MASTER'] == 0) & (data['Z_BEST'] > 0.5) 
               & (data['Z_BEST'] < 0.6) & (data['Z_USE'] < 3) 
               & (data['LAMBDAR_COSMOS2015_sep'] < 0.8))
good = buf[0]
lcbg = data[good]
n_lcbg = len(lcbg)
print n_lcbg
```
